```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt



class CloseCv:
  def __init__(self):
    pass

  def Negative(self,img):
    img = 255 - img
    return img


  def thresholding(self,img,val):
    return np.where(img>=val,255,0)

  def thresholdingINV(self,img,val):
    return np.where(img>=val,0,255)


  def clipping(self,img,v1:int,v2:int):
    img1 = np.where(img>= v1,255,img)
    img2 = np.where(img1 < v1,255,img)
    return np.where(img2==255,255,0)

  # Min- max stretching
  def min_max_stretching(self,img):
    imgM = ((img - img.min()-2)/(img.max()-2-img.min()))*255
    return imgM

  def padding(self,string):
    size = len(string)
    x='00000000'
    if size != 8:
      z = 8-size
      string = x[0:z]+string
    return string


  def bit_plane_slicing(self,img_gray):
    figure,axis = plt.subplots(2,4,figsize=(20,10))
    counter = -1
    for x in range(2):
      for y in range(4):
        counter +=1
        l=[]
        ll=[]
        for i in img_gray:
          for j in i:
            binary = bin(j)[2:]
            digit = str(padding(self,binary))
            l.append(int(digit[counter]))
          ll.append(l)
          l=[]
        msb=np.array(ll)
        axis[x,y].imshow(msb,cmap='gray')

  def log_transform(self,img):
    c=255/np.log10(256)
    img = c*np.log10(1+img)
    return img

  def power_law_trf(self,img):
    y=0.2
    imgt = img**y
    return plt.imshow(imgt,cmap='gray')


img = cv.imread('apple.jpg',0)
img1 = cv.imread('lena.jpg',0)
cc = CloseCv()

cc.power_law_trf(img1)
```
